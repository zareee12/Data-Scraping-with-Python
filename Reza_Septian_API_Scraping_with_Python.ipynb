# Day 1: API Scraping Exchange Rates

```python
# Import library
import requests

# Where USD is the base currency you want to use
# Change 'YOUR_API_KEY' with Your API KEY
url = 'https://v6.exchangerate-api.com/v6/4fabd3784facb3206672649d/latest/USD'

# Making our request
response = requests.get(url)
data = response.json()

# Your JSON object
data
```

```python
# Import library
import pandas as pd
```

```python
# Merubah json ke dataframe
data = pd.json_normalize(data)
data.head()
```

```python
# Mengecek column dan jumlahnya
data.columns
```

```python
# Mengambil 8 kolom pertama
data1 = data.iloc[:, 0:8]
data1.head()
```

```python
data1['key'] = 1
data1.head()
```

```python
data1['time_last_update_utc'] = pd.to_datetime(data['time_last_update_utc'])
data1
```

```python
# from pytz import timezone
# wib = timezone('Asia/Jakarta')
# data1['time_last_update_utc'] = data1['time_last_update_utc'].astimezone(timezone('Asia/jakarta'))
```

```python
# Mengambil kolom ke 9 - 170 untuk kemudian di transpose
data2 = data.iloc[:,8:].T
data2.head()
```

```python
# Menambah kolom baru 'key'
data2['key'] = 1
data2.head()
```

```python
# Merubah nama kolom '0' menjadi 'rate'
data2 = data2.rename(columns={0:'rate'})
data2.head()
```

```python
# Merubah index menjadi column dengan nama 'currency'
data2['currency'] = data2.index
data2.head()
```

```python
# Menghapus karakter 'conversion_rates.' di kolom currency
data2['currency'] = data2['currency'].str.replace(r'conversion_rates.', '', regex=True)
data2.head()
```

```python
# Merubah posisi kolom
data2 = data2[['key','currency','rate']]
# Menghapus index
data2.reset_index(drop=True, inplace=True)
data2.head()
```

```python
data_merged = pd.merge(data1, data2, on='key')
```

```python
data_merged.head()
```

```python
data_merged = data_merged[['time_last_update_utc','base_code','currency','rate']]
data_merged.head()
```

```python
data_merged.to_csv('exchange_rates_data.csv')
```
